```python
# Îç∞Ïù¥ÌÑ∞Îäî Ï∫êÍ∏ÄÏóêÏÑú Î∞õÏúºÏÖîÎèÑ ÎêòÍ≥† Ïù¥ ÎßÅÌÅ¨ÏóêÏÑú Î∂àÎü¨ÏôÄÎèÑ Îê©ÎãàÎã§.
import pandas as pd
df = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kc_house_data/kc_house_data.csv')
```

### 1) ÌäπÏÑ±Îì§Í≥º ÌÉÄÍ≤ü(Price)Í≥ºÏùò ÏÉÅÍ¥ÄÍ≥ÑÏàòÎ•º Í≥ÑÏÇ∞ÌïòÍ≥† Í∞ÄÏû• ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÎÜíÏùÄ ÌäπÏÑ±ÏùÑ Ï∞æÏïÑ ÌÉÄÍ≤üÍ≥ºÏùò Í¥ÄÍ≥ÑÎ•º ÏãúÍ∞ÅÌôîÌïòÏÑ∏Ïöî.
- **(Urclass Quiz) ÌÉÄÍ≤üÍ≥º ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä Í∞ÄÏû• ÎÜíÏùÄ ÌäπÏÑ±Ïùò Ïù¥Î¶ÑÏùÑ Ï†úÏ∂úÌïòÏÑ∏Ïöî.**

```python
### Ïù¥Í≥≥ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥ Ï£ºÏÑ∏Ïöî ###
# df.corrwith(df['price'], method='spearman')  
df.corr(method='pearson') # pearson ?? 
```

```python
import seaborn as sns
sns.heatmap(data = df.corr(), annot=True, fmt = '.2f', linewidths=.5, cmap='Blues')
sns.set(rc = {'figure.figsize':(10,8)})
```

```python
sns.scatterplot(x= df["sqft_living"], y = df["price"])
```

### 2) sqft_living ÏôÄ priceÏùò scatter plotÏùÑ Í∑∏Î¶¨Í≥† Í∏∞Ï§ÄÎ™®Îç∏ÏùÑ ÏãúÍ∞ÅÌôîÌï¥ Î≥¥ÏÑ∏Ïöî.
- **(Urclass Quiz) Í∏∞Ï§ÄÎ™®Îç∏Ïùò Í∞íÏùÑ Ï†úÏ∂úÌïòÏÑ∏Ïöî. (ÏÜåÏà´Ï†ê Ïù¥Ìïò Î∞òÏò¨Î¶º)**

```python
### Ïù¥Í≥≥ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥ Ï£ºÏÑ∏Ïöî ###
import matplotlib.pyplot as plt

price_mean  = df["price"].mean()
# mae Ï†àÎåÄÏóêÎü¨ÌèâÍ∑†
error = price_mean - df["price"]
print(error)

mae = error.abs().mean()

sns.scatterplot(x= df["sqft_living"], y = df["price"])
plt.axhline(mae)
print(round(price_mean)) # ÏÑ†ÌòïÌöåÍ∏∞ Í∏∞Ï§ÄÎ™®Îç∏ÏùÄ ÌÉÄÍ≤üÏùò ÌéëÍ∑†Í∞íÏù¥Îã§. 
sns.lineplot(x =df["sqft_living"], y = round(price_mean) ,color ="r")
```

### 3) Scikit-Learn ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º ÏÇ¨Ïö©Ìï¥ ÌäπÏÑ± sqft_livingÏóê ÎåÄÌïú ÏÑ†ÌòïÌöåÍ∑ÄÎ™®Îç∏ÏùÑ ÎßåÎì§Ïñ¥ Î≥¥ÏÑ∏Ïöî.
- **(Urclass Quiz) sqft_living Í∞íÏù¥ 15000Ïù∏ Í≤ΩÏö∞Ïóê Î™®Îç∏Ïù¥ ÏòàÏ∏°ÌïòÎäî Ï£ºÌÉù Í∞ÄÍ≤©ÏùÑ Ï†úÏ∂úÌïòÏÑ∏Ïöî.(ÏÜåÏàòÏ†ê Ïù¥Ìïò Î∞òÏò¨Î¶º)**

```python
### Ïù¥Í≥≥ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥ Ï£ºÏÑ∏Ïöî ###
from sklearn.linear_model import LinearRegression
import numpy as np
model = LinearRegression()
print(len(df["sqft_living"]))
df_fit = df.iloc[:15000] #ÌïôÏäµÎç∞Ïù¥ÌÑ∞ 
df_test = df["sqft_living"].iloc[15000:21613]  # ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ 
X_train = df_fit[["sqft_living"]]
y_train = df_fit[["price"]]
model.fit(X_train.values, y_train.values) # ÌïôÏäµ ÏãúÏûë 

# test data 
X = [[15000]]
result = model.predict(X)
np.round(result)

X_test = [[x] for x in df_test ]
y_pred = model.predict(X_test)

## train Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú Í∑∏ÎûòÌîÑÎ•º Í∑∏Î†§Î≥¥Í≤†ÏäµÎãàÎã§.
plt.scatter(X_train, y_train, color='black', linewidth=1)

## test Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÏòàÏ∏°ÏùÑ ÌååÎûÄÏÉâ Ï†êÏúºÎ°ú ÎÇòÌÉÄÎÇ¥ Î≥¥Í≤†ÏäµÎãàÎã§.
plt.scatter(X_test, y_pred, color='blue', linewidth=1);

print(X_train)
print(X_train.values)
```

### 4) ÎßåÎì† Î™®Îç∏ÏùÑ ÏÇ¨Ïö©Ìï¥ ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ ÏòàÏ∏°ÏùÑ Ìï¥ Î≥¥Í≥† coefficientÎ•º ÏÇ¨Ïö©Ìï¥ ÏÑ§Î™ÖÌï¥ Î≥¥ÏÑ∏Ïöî.
- **(Urclass Quiz) sqft_living ÌäπÏÑ±Ïóê ÎåÄÌïú ÌöåÍ∑ÄÍ≥ÑÏàòÎ•º Ï†úÏ∂úÌïòÏÑ∏Ïöî. (ÏÜåÏàòÏ†ê Ïù¥Ìïò Î∞òÏò¨Î¶º)**

```python
### Ïù¥Í≥≥ÏóêÏÑú Í≥ºÏ†úÎ•º ÏßÑÌñâÌï¥ Ï£ºÏÑ∏Ïöî ###
print(model.coef_) #Í≥ÑÏàò  , Í∏∞Ïö∏Í∏∞   
## Ï†àÌé∏(intercept)
model.intercept_  
```
